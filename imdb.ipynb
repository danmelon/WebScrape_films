```python
#importing required Libraries
import requests, openpyxl
import requests       #to send the request to the URL
from bs4 import BeautifulSoup #to get the content in the form of HTML


excel = openpyxl.Workbook()
sheet = excel.active
sheet.title = 'Top Rated Movies on IMDB'

sheet.append(['Film Title','Rank','Year of Release','IMDB Rating'])


#assigning the URL with variable name url
url = 'https://www.imdb.com/chart/top/?ref_=nv_mv_250'
#request allow you to send HTTP request
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

#creating an empty list, so that we can append the values

#storing the meaningfull required data in the variable
movie_data = soup.find('tbody', class_= "lister-list").find_all('tr')


#calling one by one using for loop
for store in movie_data:
    
    
    name = store.find('td', class_="titleColumn").a.text
  
    
    rank = store.find('td', class_="titleColumn").get_text(strip=True).split('.')[0]

    
    year_of_release = store.find('td', class_="titleColumn").span.text.strip('()')
 
    
    rate = store.find('td', class_="ratingColumn imdbRating").strong.text


    print(name, rank, year_of_release, rate)
    sheet.append([name, rank, year_of_release, rate])
    
excel.save('imdb movie ratings.xlsx')
```
